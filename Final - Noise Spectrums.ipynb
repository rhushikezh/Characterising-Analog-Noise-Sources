# Importing all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

# White Noise

```python
# Generate white noise
fs = 48000  # Sampling frequency
duration = 5  # seconds
samples = int(fs * duration)
np.random.seed(0)
white_noise = np.random.randn(samples)

# Estimating power spectral density using Welch's method
frequencies, psd = welch(white_noise, fs, nperseg=1024)

# Plotting the power spectral density
plt.figure(figsize=(8, 4))
plt.semilogy(frequencies, psd, linewidth = 0.5, color = 'black')
plt.title('Power Spectral Density of White Noise', fontsize = 12)
plt.xlabel('Frequency (Hz)', fontsize = 12)
plt.ylabel(r'PSD ($V^2/Hz$)', fontsize = 12)
plt.grid(True, linestyle='--', linewidth=0.3, color='gray')

# Note: Replace the file path with your actual save path
plt.savefig("/Users/goldenswan/Desktop/Thesis/Data/Plots/Spectrum/PSD_WhiteNoise.png")

plt.show()
```
