# Importing all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import butter, lfilter, decimate
```

```python
# For demonstration purposes, generating some random data
# Shall replace these with the actual data
time = np.linspace(0, 7, 500)  # Time values from 0 to 10 with 1000 points
value = np.random.rand(500)  # Random values

# Compute the Fourier transform
fft_result = np.fft.fft(value)

# Computing the frequencies corresponding to the Fourier transform
# The frequency range is from 0 to the Nyquist frequency
# Nyquist frequency = 1 / (2 * delta_t)
# delta_t is the time interval between samples
delta_t = time[1] - time[0]
frequencies = np.fft.fftfreq(len(time), delta_t)

# Plot time vs value
plt.figure(figsize=(5, 4))
plt.plot(time, value, linewidth=0.25, color = 'black')
plt.title('Time vs Value', fontsize = 12)
plt.xlabel('Time(s)', fontsize = 12)
plt.ylabel('Value (arbitrary unit)', fontsize = 12)
plt.grid(True, linestyle='--', linewidth=0.3, color='gray')

plt.savefig('/Users/goldenswan/Desktop/Thesis/Data/Plots/NoiseGen/PRNG.png')


plt.show()
```

```python
# Define a general function for downsampling
def downsample_data(time, values, downsampling_factor):
    values_downsampled = decimate(values, downsampling_factor, ftype='fir')
    time_downsampled = time[::downsampling_factor]
    return time_downsampled, values_downsampled

# Datasets
datasets = {
    '1': BJT_lp_2,
    '2': BJT_lp_2,
    '3': BJT_lp_3,
    '4': BJT_lp_4,
}

# Create a 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
axs = axs.flatten()  # Flatten to 1D array for easier access

# Define the downsampling factor
downsampling_factor = 10000

# Plot each dataset
for i, (label, dataset) in enumerate(datasets.items()):
    time_downsampled, values_downsampled = downsample_data(dataset['Time'], dataset['Value'], downsampling_factor)
    
    axs[i].plot(time_downsampled, values_downsampled, linewidth=0.3)
    axs[i].set_title(f' Dataset {label}', fontsize = 12)
    axs[i].set_xlabel('Time (ns)', fontsize = 12)
    axs[i].set_ylabel('Value (V)', fontsize = 12)
    axs[i].grid(True, linestyle='--', linewidth=0.3, color='gray')
    
plt.tight_layout()

plt.savefig('/Users/goldenswan/Desktop/Thesis/Data/Plots/NoiseGen/BJT_LP.png')
plt.show()
```
